```python
from bs4 import BeautifulSoup  # For HTML parsing
import requests                # To get web pages
```

## Wall Street Journal

```python
wsj = requests.get('http://www.wsj.com/')
```

```python
wsj_soup = BeautifulSoup(wsj.text, 'html.parser')
print(wsj_soup.prettify())
```

```python
wsj_soup.find_all("h3")
```

### Getting Headline  Text

```python
wsj_soup.find_all("h3")[0].string
```

```python
# From a BeautifulSoup h3 object, return string
def get_string(obj):
    return obj.string

wsj_headlines = [get_string(headline) for headline in wsj_soup.find_all("h3")]
wsj_headlines
```

```python
wsj_headlines_text = " ".join([headline for headline in wsj_headlines if headline])
wsj_headlines_text
```

## New York Times


```python
nyt = requests.get('http://www.nytimes.com/')
nyt_soup = BeautifulSoup(nyt.text, 'html.parser')
print(nyt_soup.prettify())
```

## Grab Headlines

```python
nyt_soup.find_all("h1")
```

```python
nyt_soup.find_all("h2")
```

```python
nyt_soup.find_all("h3")
```

It seems like finding all tags with the class "story-heading" will be more efficient than grabbing all header tags.

```python
def nyt_heading(tag):
    try:
        # Some tags have multiple classes
        if 'story-heading' in tag['class']:
            return True
        
    # Not every tag has a class attribute
    except KeyError:
        return False
    
    return False

nyt_headlines_raw = nyt_soup.find_all(nyt_heading)
nyt_headlines_raw[0:5]
```

```python
nyt_headlines_raw[0].string
```

```python
nyt_headlines = [get_string(headline) for headline in nyt_headlines_raw]
nyt_headlines
```

```python
nyt_headlines_text = " ".join([headline.replace("\n", "") for headline in nyt_headlines if headline])
nyt_headlines_text
```

## National Public Radio


```python
npr = requests.get("http://www.npr.org/")
npr_soup = BeautifulSoup(npr.text, "html.parser")
print(npr_soup.prettify())
```

```python
npr_soup.find_all("h1")
```

```python
npr_soup.find_all("h2")
```

```python
def npr_heading(tag):
    try:
        # Some tags have multiple classes
        if 'title' in tag['class']:
            return True
        
    # Not every tag has a class attribute
    except KeyError:
        return False
    
    return False

npr_headlines_raw = npr_soup.find_all(npr_heading)
npr_headlines_raw
```

```python
npr_headlines = [get_string(headline) for headline in npr_headlines_raw]
npr_headlines
```

```python
npr_headlines_text = " ".join(npr_headlines)
npr_headlines_text
```

## The Atlantic

```python
atl = requests.get("https://www.theatlantic.com")
atl_soup = BeautifulSoup(atl.text, "html.parser")
print(atl_soup.prettify())
```

```python
atl_soup.find_all("h1")
```

```python
def atl_heading(tag):
    try:
        # Some tags have multiple classes
        if 'hed' in tag['class']:
            return True
        
    # Not every tag has a class attribute
    except KeyError:
        return False
    
    return False

atl_headlines_raw = atl_soup.find_all(atl_heading)
atl_headlines_raw
```

```python
atl_headlines = [get_string(headline) for headline in atl_headlines_raw if get_string(headline)]
atl_headlines
```

```python
atl_headlines_text = " ".join(atl_headlines)
atl_headlines_text
```

## Word Cloud

```python
from wordcloud import WordCloud

news_wc = WordCloud(width=800, height=600, background_color="white").generate(
    npr_headlines_text + nyt_headlines_text + wsj_headlines_text + atl_headlines_text)
```

```python
news_wc.to_file('news.jpg')
```
